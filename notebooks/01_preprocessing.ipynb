```python
# 📚 Import all required libraries
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

print("✅ All libraries imported successfully!")
print(f"📊 Pandas version: {pd.__version__}")
print(f"🔢 NumPy version: {np.__version__}")
```

```python
# 📂 Load your 2 lakh rows dataset
dataset_path = '../DATASET/ecommerce.csv'

print("⏳ Loading dataset... This may take a moment for 2 lakh rows...")
df = pd.read_csv(dataset_path)

print(f"\n✅ Dataset loaded successfully!")
print(f"📊 Total rows: {len(df):,}")
print(f"📋 Total columns: {len(df.columns)}")
print(f"\n📌 Column names: {df.columns.tolist()}")
```

```python
# 👀 Preview first 5 rows of your dataset
print("=" * 80)
print("🔍 FIRST 5 ROWS OF YOUR DATASET:")
print("=" * 80)
df.head()
```

```python
# 📊 Get dataset info
print("=" * 80)
print("📊 DATASET INFORMATION:")
print("=" * 80)
df.info()

print("\n" + "=" * 80)
print("📈 STATISTICAL SUMMARY:")
print("=" * 80)
df.describe()
```

```python
# 🔍 Check for missing values
print("=" * 80)
print("❓ MISSING VALUES CHECK:")
print("=" * 80)

missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100

missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Missing %': missing_pct.round(2)
})

print(missing_df)
print(f"\n📊 Total missing values: {missing.sum():,}")
```

```python
# 🧹 Clean and enhance the dataset
print("=" * 80)
print("🧹 CLEANING & ENHANCING THE DATASET...")
print("=" * 80)

original_rows = len(df)

# Step 1: Convert data types properly
print("\n1️⃣ Converting data types...")
df['Original_Price'] = pd.to_numeric(df['Original_Price'], errors='coerce')
df['Final_Price'] = pd.to_numeric(df['Final_Price'], errors='coerce')
df['Discount_Percentage'] = pd.to_numeric(df['Discount_Percentage'], errors='coerce')
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')
df['Review_Count'] = pd.to_numeric(df['Review_Count'], errors='coerce')
print("   ✅ All numeric columns converted properly")

# Step 2: Fill any values that couldn't be converted
print("\n2️⃣ Handling converted values...")
df['Original_Price'] = df['Original_Price'].fillna(df['Original_Price'].median())
df['Final_Price'] = df['Final_Price'].fillna(df['Final_Price'].median())
df['Discount_Percentage'] = df['Discount_Percentage'].fillna(0)
df['Rating'] = df['Rating'].fillna(df['Rating'].median())
df['Review_Count'] = df['Review_Count'].fillna(0).astype(int)
print("   ✅ Missing values after conversion handled")

# Step 3: Remove duplicates
print("\n3️⃣ Checking for duplicates...")
before = len(df)
df = df.drop_duplicates()
removed = before - len(df)
print(f"   ✅ Removed {removed:,} duplicate rows")
print(f"   📊 Rows remaining: {len(df):,}")

# Step 4: Clean text fields
print("\n4️⃣ Cleaning text fields...")
df['Category'] = df['Category'].astype(str).str.strip()
df['vendors'] = df['vendors'].astype(str).str.strip()
df['product_name'] = df['product_name'].astype(str).str.strip()
df['Customer_Review'] = df['Customer_Review'].astype(str).str.strip()
print("   ✅ Text fields cleaned")

# Step 5: Create new features for ML
print("\n5️⃣ Creating new features...")

# Price Difference
df['Price_Difference'] = df['Original_Price'] - df['Final_Price']
print("   ✅ Price_Difference calculated")

# Savings Ratio
df['Savings_Ratio'] = (df['Price_Difference'] / df['Original_Price']).round(4)
df['Savings_Ratio'] = df['Savings_Ratio'].fillna(0)
print("   ✅ Savings_Ratio calculated")

# Price Category (Budget/Mid-Range/Premium/Luxury)
df['Price_Category'] = pd.cut(
    df['Final_Price'], 
    bins=[0, 500, 2000, 10000, float('inf')],
    labels=['Budget', 'Mid-Range', 'Premium', 'Luxury'],
    include_lowest=True
)
print("   ✅ Price_Category created (Budget/Mid-Range/Premium/Luxury)")

# Rating Category
df['Rating_Category'] = pd.cut(
    df['Rating'],
    bins=[-0.1, 2, 3.5, 4.5, 5.1],
    labels=['Poor', 'Average', 'Good', 'Excellent'],
    include_lowest=True
)
print("   ✅ Rating_Category created (Poor/Average/Good/Excellent)")

print(f"\n{'=' * 80}")
print(f"✅ CLEANING COMPLETE!")
print(f"📊 Original rows: {original_rows:,}")
print(f"📊 Final rows: {len(df):,}")
print(f"📋 Total columns now: {len(df.columns)} (added 4 new features!)")
print(f"{'=' * 80}")

# Show all columns
print(f"\n📌 All columns now:")
for i, col in enumerate(df.columns, 1):
    print(f"   {i}. {col} ({df[col].dtype})")
```

```python
# 💾 Save the cleaned dataset
output_path = '../DATASET/cleaned_dataset.csv'

print("=" * 80)
print("💾 SAVING CLEANED DATASET...")
print("=" * 80)

df.to_csv(output_path, index=False)

print(f"\n✅ Saved successfully!")
print(f"📁 Location: {output_path}")
print(f"📊 Total rows saved: {len(df):,}")
print(f"📋 Total columns: {len(df.columns)}")

# Verify by reading it back
print(f"\n🔍 Verifying saved file...")
df_verify = pd.read_csv(output_path)
print(f"   ✅ File reads correctly")
print(f"   📊 Verified rows: {len(df_verify):,}")

print("\n" + "=" * 80)
print("🔍 PREVIEW OF CLEANED DATA:")
print("=" * 80)
df.head()
```

```python
# 🔧 FIX: Remove/Cap outliers for realistic prices
print("=" * 80)
print("🔧 FIXING OUTLIERS - CAPPING PRICES AT REALISTIC VALUES")
print("=" * 80)

before_rows = len(df)

# Set realistic maximum price (₹10 lakhs = ₹10,00,000)
MAX_REALISTIC_PRICE = 1000000  # 10 Lakhs

print(f"\n📊 Setting maximum realistic price: ₹{MAX_REALISTIC_PRICE:,}")

# Count outliers
outliers = len(df[df['Final_Price'] > MAX_REALISTIC_PRICE])
print(f"🔍 Found {outliers} products above ₹{MAX_REALISTIC_PRICE:,}")

# Option 1: REMOVE outliers (cleaner approach)
df = df[df['Final_Price'] <= MAX_REALISTIC_PRICE]
df = df[df['Original_Price'] <= MAX_REALISTIC_PRICE * 1.5]  # Original can be 50% higher

# Also remove any weird negative or zero prices
df = df[df['Final_Price'] > 0]
df = df[df['Original_Price'] > 0]

after_rows = len(df)
removed = before_rows - after_rows

print(f"\n✅ FIXED!")
print(f"📊 Removed {removed} outlier rows")
print(f"📊 Remaining rows: {after_rows:,}")

# Show new statistics
print(f"\n💰 NEW PRICE STATISTICS (Cleaned):")
print(f"   Min Price: ₹{df['Final_Price'].min():,.2f}")
print(f"   Max Price: ₹{df['Final_Price'].max():,.2f}")
print(f"   Mean Price: ₹{df['Final_Price'].mean():,.2f}")
print(f"   Median Price: ₹{df['Final_Price'].median():,.2f}")

# Recalculate features
print(f"\n🔄 Recalculating features...")
df['Price_Difference'] = df['Original_Price'] - df['Final_Price']
df['Savings_Ratio'] = (df['Price_Difference'] / df['Original_Price']).round(4)
df['Savings_Ratio'] = df['Savings_Ratio'].fillna(0)

# Recalculate Price_Category
df['Price_Category'] = pd.cut(
    df['Final_Price'], 
    bins=[0, 500, 2000, 10000, float('inf')],
    labels=['Budget', 'Mid-Range', 'Premium', 'Luxury'],
    include_lowest=True
)

print(f"\n🎯 Now showing REALISTIC data!")
print(f"\n🔝 TOP 5 MOST EXPENSIVE PRODUCTS NOW:")
print(df.nlargest(5, 'Final_Price')[['product_name', 'Category', 'Final_Price', 'vendors']])
```
