```python
# 📊 Import visualization libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings
warnings.filterwarnings('ignore')

# Set style for matplotlib
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

# Configure plotly
import plotly.io as pio
pio.templates.default = "plotly_white"

print("✅ Visualization libraries imported!")
print("📊 Ready to create beautiful graphs!")
print(f"🎨 Matplotlib version: {plt.matplotlib.__version__}")
print(f"🌊 Seaborn version: {sns.__version__}")
```

```python
# 📂 Load dataset with sentiment analysis
dataset_path = '../DATASET/dataset_with_sentiment.csv'

print("⏳ Loading dataset...")
df = pd.read_csv(dataset_path)

print(f"\n✅ Dataset loaded!")
print(f"📊 Total rows: {len(df):,}")
print(f"📋 Total columns: {len(df.columns)}")

# Quick summary
print("\n" + "=" * 60)
print("📋 DATASET OVERVIEW:")
print("=" * 60)
print(f"📁 Categories: {df['Category'].nunique()}")
print(f"🏪 Vendors: {df['vendors'].nunique()}")
print(f"🎭 Positive Reviews: {len(df[df['Sentiment_Label']=='POSITIVE']):,}")
print(f"🎭 Negative Reviews: {len(df[df['Sentiment_Label']=='NEGATIVE']):,}")
print(f"💰 Price Range: ₹{df['Final_Price'].min():.2f} - ₹{df['Final_Price'].max():.2f}")
print(f"⭐ Avg Rating: {df['Rating'].mean():.2f}")

print("\n🎨 Ready to create visualizations!")
```

```python
# 🎭 Sentiment Distribution - Pie Chart
print("=" * 80)
print("🎭 SENTIMENT DISTRIBUTION PIE CHART")
print("=" * 80)

# Calculate sentiment counts
sentiment_counts = df['Sentiment_Label'].value_counts()
colors = ['#10B981', '#F59E0B', '#EF4444']  # Green, Orange, Red

# Create pie chart with Plotly (interactive)
fig = px.pie(
    values=sentiment_counts.values,
    names=sentiment_counts.index,
    title=f"<b>Customer Sentiment Distribution</b><br><sup>{len(df):,} Reviews Analyzed</sup>",
    color_discrete_sequence=colors,
    hole=0.4,  # Donut chart
    width=800,
    height=500
)

# Customize layout
fig.update_traces(
    textposition='inside',
    textinfo='percent+label',
    textfont_size=16,
    marker_line_color='white',
    marker_line_width=3
)

fig.update_layout(
    font_family="Arial Black",
    font_size=14,
    title_font_size=24,
    showlegend=True,
    legend_font_size=12
)

# Show interactive chart
fig.show()

# Print summary
print(f"\n📊 Sentiment Summary:")
print(f"🟢 POSITIVE: {sentiment_counts.get('POSITIVE', 0):,} ({sentiment_counts.get('POSITIVE', 0)/len(df)*100:.2f}%)")
print(f"🔴 NEGATIVE: {sentiment_counts.get('NEGATIVE', 0):,} ({sentiment_counts.get('NEGATIVE', 0)/len(df)*100:.2f}%)")
print(f"\n💡 Insight: {sentiment_counts.get('POSITIVE', 0)/len(df)*100:.1f}% customers are happy!")
```

```python
# 💰 Average Price by Category - CLEAN VERSION
print("=" * 80)
print("💰 AVERAGE PRICE BY CATEGORY (Main Categories Only)")
print("=" * 80)

# Define the REAL categories we want to show
main_categories = [
    'Electronics', 
    'Grocery', 
    'Snacks', 
    'Beverages', 
    'Fruits & Vegetables', 
    'Fashion', 
    'Home & Living', 
    'Beauty',
    'Books',
    'Clothing'
]

# Filter dataset to only main categories
df_clean = df[df['Category'].isin(main_categories)].copy()

print(f"\n📊 Filtering dataset:")
print(f"   Original rows: {len(df):,}")
print(f"   Clean rows (main categories only): {len(df_clean):,}")

# Calculate average price by category
category_avg_price = df_clean.groupby('Category')['Final_Price'].agg(['mean', 'count']).reset_index()
category_avg_price.columns = ['Category', 'Average_Price', 'Product_Count']
category_avg_price = category_avg_price.sort_values('Average_Price', ascending=False)

# Create bar chart
fig = px.bar(
    category_avg_price,
    x='Category',
    y='Average_Price',
    title="<b>Average Product Price by Category</b><br><sup>Pricing patterns across main product categories</sup>",
    color='Average_Price',
    color_continuous_scale='Viridis',
    text='Average_Price',
    width=1000,
    height=600,
    hover_data={'Product_Count': True, 'Average_Price': ':.2f'}
)

# Customize
fig.update_traces(
    texttemplate='₹%{text:.0f}',
    textposition='outside',
    textfont_size=14,
    marker_line_color='white',
    marker_line_width=2
)

fig.update_layout(
    font_family="Arial",
    font_size=14,
    title_font_size=22,
    xaxis_title="<b>Category</b>",
    yaxis_title="<b>Average Price (₹)</b>",
    xaxis_tickangle=-45,
    showlegend=False,
    plot_bgcolor='white'
)

fig.show()

# Print insights
print(f"\n📊 Category Pricing Insights:")
print("-" * 60)
for idx, row in category_avg_price.iterrows():
    print(f"   📁 {row['Category']:25} Avg: ₹{row['Average_Price']:10,.2f} ({row['Product_Count']:,} products)")

print(f"\n💡 Most Expensive: {category_avg_price.iloc[0]['Category']} (₹{category_avg_price.iloc[0]['Average_Price']:,.2f})")
print(f"💡 Most Affordable: {category_avg_price.iloc[-1]['Category']} (₹{category_avg_price.iloc[-1]['Average_Price']:,.2f})")
print(f"📦 Total products shown: {category_avg_price['Product_Count'].sum():,}")
```

```python
# ⭐ Rating vs Price Scatter Plot - FIXED VERSION
print("=" * 80)
print("⭐ RATING vs PRICE RELATIONSHIP")
print("=" * 80)

# Filter to realistic prices (below ₹10,000 for better visualization)
df_sample = df_clean[df_clean['Final_Price'] <= 10000].copy()
df_sample = df_sample.sample(min(5000, len(df_sample)), random_state=42)

print(f"\n📊 Plotting {len(df_sample):,} products (prices under ₹10,000)")

# Create scatter plot
fig = px.scatter(
    df_sample,
    x='Final_Price',
    y='Rating',
    color='Category',
    size='Review_Count',
    hover_data=['product_name', 'vendors', 'Discount_Percentage'],
    title="<b>Rating vs Price Analysis</b><br><sup>Products under ₹10,000 | Dot size = Review count</sup>",
    labels={
        'Final_Price': 'Price (₹)',
        'Rating': 'Customer Rating',
        'Review_Count': 'Review Count'
    },
    width=1200,
    height=700,
    opacity=0.7,
    size_max=30
)

# Set axis ranges
fig.update_layout(
    font_family="Arial",
    font_size=14,
    title_font_size=22,
    xaxis_title="<b>Price (₹)</b>",
    yaxis_title="<b>Rating (1-5 stars)</b>",
    plot_bgcolor='white',
    hovermode='closest',
    xaxis=dict(range=[0, 10000], gridcolor='lightgray'),
    yaxis=dict(range=[0, 5.5], gridcolor='lightgray')
)

# Add average rating line
avg_rating = df_sample['Rating'].mean()
fig.add_hline(
    y=avg_rating, 
    line_dash="dash", 
    line_color="red",
    annotation_text=f"Avg Rating: {avg_rating:.2f}",
    annotation_position="top right"
)

fig.show()

# Calculate correlation
correlation = df_clean['Final_Price'].corr(df_clean['Rating'])

print(f"\n📊 STATISTICAL ANALYSIS:")
print("-" * 50)
print(f"   📈 Correlation (Price vs Rating): {correlation:.4f}")
print(f"   ⭐ Average Rating: {df_sample['Rating'].mean():.2f}")
print(f"   💰 Average Price: ₹{df_sample['Final_Price'].mean():.2f}")

print(f"\n💡 Products Rating Distribution:")
rating_dist = df_sample['Rating'].apply(lambda x: int(x)).value_counts().sort_index()
for rating, count in rating_dist.items():
    bar = '█' * int(count/100)
    print(f"   {rating}⭐: {count:,} products {bar}")

# Interpret correlation
if correlation > 0.5:
    interpretation = "Strong positive - Higher price = Better rating"
elif correlation > 0.2:
    interpretation = "Weak positive - Slight tendency for higher priced items to have better ratings"
elif correlation > -0.2:
    interpretation = "No clear relationship - Price doesn't affect rating much"
else:
    interpretation = "Negative - Lower prices have better ratings"

print(f"\n💡 INTERPRETATION: {interpretation}")
```

```python
# 🏷️ Discount vs Rating Heatmap
print("=" * 80)
print("🏷️ DISCOUNT vs RATING HEATMAP")
print("=" * 80)

# Create rating bins
df_clean['Rating_Bin'] = pd.cut(df_clean['Rating'], 
                                  bins=[0, 2, 3, 4, 5],
                                  labels=['1-2 ⭐', '2-3 ⭐', '3-4 ⭐', '4-5 ⭐'])

# Create discount bins
df_clean['Discount_Bin'] = pd.cut(df_clean['Discount_Percentage'], 
                                    bins=[0, 10, 25, 40, 60, 100],
                                    labels=['0-10%', '10-25%', '25-40%', '40-60%', '60%+'])

# Create pivot table
heatmap_data = df_clean.groupby(['Rating_Bin', 'Discount_Bin']).size().reset_index(name='Count')
heatmap_pivot = heatmap_data.pivot(index='Rating_Bin', columns='Discount_Bin', values='Count').fillna(0)

# Create heatmap
fig = px.imshow(
    heatmap_pivot,
    labels=dict(x="Discount Range", y="Rating Range", color="Product Count"),
    title="<b>Rating vs Discount Heatmap</b><br><sup>How ratings correlate with discount percentages</sup>",
    color_continuous_scale='RdYlGn',
    aspect='auto',
    width=1000,
    height=600,
    text_auto=True
)

fig.update_layout(
    font_family="Arial",
    font_size=14,
    title_font_size=22,
    xaxis_title="<b>Discount Range</b>",
    yaxis_title="<b>Customer Rating</b>"
)

fig.show()

print(f"\n📊 Heatmap Analysis:")
print("-" * 50)
print(heatmap_pivot.astype(int))

print(f"\n💡 Key Insights:")
# Find cells with most products
max_cell = heatmap_data.loc[heatmap_data['Count'].idxmax()]
print(f"   🎯 Most common combo: {max_cell['Rating_Bin']} rating + {max_cell['Discount_Bin']} discount")
print(f"   📦 Products in this combo: {max_cell['Count']:,}")
```

```python
# 🏪 Top Vendors Analysis
print("=" * 80)
print("🏪 TOP VENDORS BY PRODUCT COUNT & RATING")
print("=" * 80)

# Calculate vendor statistics
vendor_stats = df_clean.groupby('vendors').agg({
    'product_name': 'count',
    'Rating': 'mean',
    'Final_Price': 'mean',
    'Discount_Percentage': 'mean',
    'Sentiment_Polarity': 'mean'
}).reset_index()

vendor_stats.columns = ['Vendor', 'Product_Count', 'Avg_Rating', 
                         'Avg_Price', 'Avg_Discount', 'Avg_Sentiment']
vendor_stats = vendor_stats.sort_values('Product_Count', ascending=False)

# Create grouped bar chart
fig = go.Figure()

fig.add_trace(go.Bar(
    x=vendor_stats['Vendor'],
    y=vendor_stats['Product_Count'],
    name='Product Count',
    marker_color='#EC4899',
    text=vendor_stats['Product_Count'],
    textposition='outside',
    texttemplate='%{text:,}'
))

fig.update_layout(
    title="<b>Top Vendors by Product Count</b><br><sup>Showing vendor dominance in the marketplace</sup>",
    xaxis_title="<b>Vendor</b>",
    yaxis_title="<b>Number of Products</b>",
    width=1200,
    height=600,
    font_family="Arial",
    font_size=14,
    title_font_size=22,
    plot_bgcolor='white',
    xaxis_tickangle=-45
)

fig.show()

# Print detailed vendor analysis
print(f"\n📊 VENDOR DETAILED ANALYSIS:")
print("-" * 100)
print(f"{'Vendor':<20} {'Products':<12} {'Avg Rating':<12} {'Avg Price':<15} {'Avg Discount':<15} {'Sentiment':<12}")
print("-" * 100)

for idx, row in vendor_stats.iterrows():
    sentiment_emoji = "🟢" if row['Avg_Sentiment'] > 0.3 else "🟡" if row['Avg_Sentiment'] > 0.1 else "🔴"
    print(f"{row['Vendor']:<20} {row['Product_Count']:<12,} {row['Avg_Rating']:<12.2f} "
          f"₹{row['Avg_Price']:<14,.2f} {row['Avg_Discount']:<14.2f}% {sentiment_emoji} {row['Avg_Sentiment']:<.3f}")

print(f"\n💡 Top Insights:")
print(f"   🏆 Biggest Vendor: {vendor_stats.iloc[0]['Vendor']} ({vendor_stats.iloc[0]['Product_Count']:,} products)")
print(f"   ⭐ Highest Rated: {vendor_stats.loc[vendor_stats['Avg_Rating'].idxmax(), 'Vendor']} "
      f"({vendor_stats['Avg_Rating'].max():.2f}⭐)")
print(f"   🎭 Best Sentiment: {vendor_stats.loc[vendor_stats['Avg_Sentiment'].idxmax(), 'Vendor']} "
      f"({vendor_stats['Avg_Sentiment'].max():.3f})")
```

```python
# 📝 Review Count Distribution
print("=" * 80)
print("📝 REVIEW COUNT DISTRIBUTION ANALYSIS")
print("=" * 80)

# Create review count bins
df_clean['Review_Bin'] = pd.cut(df_clean['Review_Count'], 
                                  bins=[0, 100, 500, 1000, 5000, float('inf')],
                                  labels=['0-100', '100-500', '500-1K', '1K-5K', '5K+'])

review_distribution = df_clean['Review_Bin'].value_counts().sort_index()

# Create histogram
fig = px.histogram(
    df_clean[df_clean['Review_Count'] <= 5000],  # Filter for better visibility
    x='Review_Count',
    nbins=50,
    title="<b>Review Count Distribution</b><br><sup>How many reviews products typically have</sup>",
    color_discrete_sequence=['#8B5CF6'],
    width=1100,
    height=600
)

fig.update_layout(
    font_family="Arial",
    font_size=14,
    title_font_size=22,
    xaxis_title="<b>Number of Reviews</b>",
    yaxis_title="<b>Product Count</b>",
    plot_bgcolor='white',
    bargap=0.1
)

fig.update_traces(
    marker_line_color='white',
    marker_line_width=1.5
)

fig.show()

# Print insights
print(f"\n📊 REVIEW COUNT INSIGHTS:")
print("-" * 60)
for bin_name, count in review_distribution.items():
    percentage = (count / len(df_clean)) * 100
    bar = '█' * int(percentage / 2)
    print(f"   {bin_name:10} reviews: {count:7,} products ({percentage:5.2f}%) {bar}")

print(f"\n💡 KEY STATISTICS:")
print(f"   📊 Total products: {len(df_clean):,}")
print(f"   📝 Average reviews: {df_clean['Review_Count'].mean():.0f}")
print(f"   📝 Median reviews: {df_clean['Review_Count'].median():.0f}")
print(f"   📝 Max reviews: {df_clean['Review_Count'].max():,}")
print(f"   📝 Min reviews: {df_clean['Review_Count'].min()}")

# Most reviewed products
print(f"\n🔥 TOP 5 MOST REVIEWED PRODUCTS:")
top_reviewed = df_clean.nlargest(5, 'Review_Count')[['product_name', 'Category', 'Review_Count', 'Rating']]
for idx, row in top_reviewed.iterrows():
    product = row['product_name'][:40] + '...' if len(str(row['product_name'])) > 40 else row['product_name']
    print(f"   📦 {product}")
    print(f"      Category: {row['Category']} | Reviews: {row['Review_Count']:,} | Rating: {row['Rating']}⭐")
```

```python
# 💰 Price Savings Analysis
print("=" * 80)
print("💰 PRICE SAVINGS ANALYSIS - HOW MUCH CUSTOMERS SAVE")
print("=" * 80)

# Calculate savings by category
savings_by_category = df_clean.groupby('Category').agg({
    'Price_Difference': 'mean',
    'Discount_Percentage': 'mean',
    'Final_Price': 'mean',
    'product_name': 'count'
}).reset_index()

savings_by_category.columns = ['Category', 'Avg_Savings', 'Avg_Discount%', 'Avg_Final_Price', 'Product_Count']
savings_by_category = savings_by_category.sort_values('Avg_Savings', ascending=False)

# Create dual-axis chart
fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=('Average Savings (₹)', 'Average Discount (%)'),
    specs=[[{'type': 'bar'}, {'type': 'bar'}]]
)

# Savings chart
fig.add_trace(
    go.Bar(
        x=savings_by_category['Category'],
        y=savings_by_category['Avg_Savings'],
        name='Savings (₹)',
        marker_color='#10B981',
        text=savings_by_category['Avg_Savings'].round(0),
        texttemplate='₹%{text}',
        textposition='outside'
    ),
    row=1, col=1
)

# Discount chart
fig.add_trace(
    go.Bar(
        x=savings_by_category['Category'],
        y=savings_by_category['Avg_Discount%'],
        name='Discount (%)',
        marker_color='#F59E0B',
        text=savings_by_category['Avg_Discount%'].round(1),
        texttemplate='%{text}%',
        textposition='outside'
    ),
    row=1, col=2
)

fig.update_layout(
    title_text="<b>💰 Customer Savings by Category</b><br><sup>Which categories offer the best deals?</sup>",
    title_font_size=22,
    width=1400,
    height=600,
    showlegend=False,
    plot_bgcolor='white'
)

fig.update_xaxes(tickangle=-45)

fig.show()

# Print insights
print(f"\n📊 SAVINGS ANALYSIS:")
print("-" * 80)
print(f"{'Category':<25} {'Avg Savings':<15} {'Avg Discount':<15} {'Avg Price':<15}")
print("-" * 80)

for idx, row in savings_by_category.iterrows():
    print(f"{row['Category']:<25} ₹{row['Avg_Savings']:<14,.2f} {row['Avg_Discount%']:<14.2f}% ₹{row['Avg_Final_Price']:<14,.2f}")

print(f"\n💡 KEY INSIGHTS:")
print(f"   🏆 Biggest Savings: {savings_by_category.iloc[0]['Category']} (₹{savings_by_category.iloc[0]['Avg_Savings']:,.2f})")
print(f"   💸 Highest Discount: {savings_by_category.loc[savings_by_category['Avg_Discount%'].idxmax(), 'Category']} "
      f"({savings_by_category['Avg_Discount%'].max():.2f}%)")
print(f"   🎯 Most Affordable: {savings_by_category.loc[savings_by_category['Avg_Final_Price'].idxmin(), 'Category']} "
      f"(₹{savings_by_category['Avg_Final_Price'].min():.2f} avg)")

total_savings = df_clean['Price_Difference'].sum()
print(f"\n💰 TOTAL CUSTOMER SAVINGS across all products: ₹{total_savings:,.2f}")
```
