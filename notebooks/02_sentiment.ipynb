```python
# 📚 Import libraries for sentiment analysis
import pandas as pd
import numpy as np
from textblob import TextBlob
import warnings
warnings.filterwarnings('ignore')

print("✅ Libraries imported successfully!")
print("🎭 Ready for sentiment analysis!")
print(f"📊 Pandas version: {pd.__version__}")
print(f"🔢 NumPy version: {np.__version__}")
```

```python
# 📂 Load the cleaned dataset (188,301 rows)
dataset_path = '../DATASET/cleaned_dataset.csv'

print("⏳ Loading cleaned dataset...")
df = pd.read_csv(dataset_path)

print(f"\n✅ Dataset loaded successfully!")
print(f"📊 Total rows: {len(df):,}")
print(f"📋 Total columns: {len(df.columns)}")
print(f"\n📌 Columns: {list(df.columns)}")

# Preview
print("\n" + "=" * 80)
print("🔍 PREVIEW (first 3 rows):")
print("=" * 80)
df[['product_name', 'Category', 'Final_Price', 'Rating', 'Customer_Review']].head(3)
```

```python
# 🎭 Test sentiment analysis on sample reviews
print("=" * 80)
print("🎭 TESTING SENTIMENT ANALYSIS")
print("=" * 80)

# Sample reviews from your dataset
sample_reviews = [
    "Amazing product! Highly recommended",
    "Too expensive for what it is",
    "Average quality, nothing special",
    "Best purchase ever! Love it!",
    "Poor quality, not worth the money",
    "Good product at good price",
    "No review"
]

print("📝 Sample reviews and their sentiment:")
print("-" * 50)

for i, review in enumerate(sample_reviews, 1):
    blob = TextBlob(review)
    sentiment = blob.sentiment.polarity
    
    if sentiment > 0.1:
        label = "🟢 POSITIVE"
    elif sentiment < -0.1:
        label = "🔴 NEGATIVE"
    else:
        label = "🟡 NEUTRAL"
    
    print(f"{i}. '{review}' → {label} (score: {sentiment:.3f})")

print("\n" + "=" * 80)
print("✅ Sentiment analysis working perfectly!")
print("📊 Polarity score: -1.0 (negative) to +1.0 (positive)")
print("🎯 Ready to analyze all 188,301 reviews!")
```

```python
# 🎭 Analyze ALL 188,301 customer reviews
print("=" * 80)
print("🎭 ANALYZING ALL 188,301 CUSTOMER REVIEWS")
print("=" * 80)

print(f"📊 Starting analysis on {len(df):,} reviews...")
print("⏳ This will take 2-5 minutes...")

# Add sentiment columns
df['Sentiment_Polarity'] = 0.0
df['Sentiment_Label'] = 'NEUTRAL'

# Analyze each review
for idx, row in df.iterrows():
    review = str(row['Customer_Review'])
    
    if review.lower() == 'no review':
        polarity = 0.0
        label = 'NEUTRAL'
    else:
        blob = TextBlob(review)
        polarity = blob.sentiment.polarity
        
        if polarity > 0.1:
            label = 'POSITIVE'
        elif polarity < -0.1:
            label = 'NEGATIVE'
        else:
            label = 'NEUTRAL'
    
    df.at[idx, 'Sentiment_Polarity'] = polarity
    df.at[idx, 'Sentiment_Label'] = label
    
    # Progress indicator (every 10,000 rows)
    if (idx + 1) % 10000 == 0:
        print(f"   📈 Processed {idx + 1:,} reviews...")

print(f"\n✅ ANALYSIS COMPLETE!")
print(f"📊 Total reviews analyzed: {len(df):,}")
print(f"📈 New columns added: Sentiment_Polarity, Sentiment_Label")
```

```python
# 📊 Show sentiment distribution
print("=" * 80)
print("📊 SENTIMENT DISTRIBUTION ACROSS 188,301 REVIEWS")
print("=" * 80)

# Count sentiments
sentiment_counts = df['Sentiment_Label'].value_counts()
total = len(df)

print("\n🎭 How customers feel about products:\n")

for sentiment, count in sentiment_counts.items():
    percentage = (count / total) * 100
    
    # Emoji based on sentiment
    if sentiment == 'POSITIVE':
        emoji = "🟢"
        bar_color = "█"
    elif sentiment == 'NEGATIVE':
        emoji = "🔴"
        bar_color = "█"
    else:
        emoji = "🟡"
        bar_color = "█"
    
    # Visual bar
    bar_length = int(percentage / 2)
    bar = bar_color * bar_length
    
    print(f"{emoji} {sentiment:10} : {count:>7,} ({percentage:5.2f}%) {bar}")

print("\n" + "=" * 80)
print("📈 POLARITY SCORE STATISTICS:")
print("=" * 80)
print(f"   Minimum: {df['Sentiment_Polarity'].min():.4f}")
print(f"   Maximum: {df['Sentiment_Polarity'].max():.4f}")
print(f"   Average: {df['Sentiment_Polarity'].mean():.4f}")
print(f"   Median:  {df['Sentiment_Polarity'].median():.4f}")

# Overall sentiment
avg_polarity = df['Sentiment_Polarity'].mean()
if avg_polarity > 0.1:
    overall = "🟢 POSITIVE"
elif avg_polarity < -0.1:
    overall = "🔴 NEGATIVE"
else:
    overall = "🟡 NEUTRAL"

print(f"\n🎯 OVERALL CUSTOMER SENTIMENT: {overall}")
print(f"💡 Average polarity: {avg_polarity:.4f}")
```

```python
# 💾 Save dataset with sentiment analysis
output_path = '../DATASET/dataset_with_sentiment.csv'

print("=" * 80)
print("💾 SAVING DATASET WITH SENTIMENT ANALYSIS")
print("=" * 80)

# Save
df.to_csv(output_path, index=False)

print(f"\n✅ Saved successfully!")
print(f"📁 Location: {output_path}")
print(f"📊 Total rows: {len(df):,}")
print(f"📋 Total columns: {len(df.columns)}")

# Verify
print(f"\n🔍 Verifying saved file...")
df_verify = pd.read_csv(output_path)
print(f"   ✅ File reads correctly: {len(df_verify):,} rows")

# Show new columns
print(f"\n📌 All columns now ({len(df.columns)} total):")
for i, col in enumerate(df.columns, 1):
    if col in ['Sentiment_Polarity', 'Sentiment_Label']:
        print(f"   {i:2}. {col} ⭐ NEW!")
    else:
        print(f"   {i:2}. {col}")

print("\n" + "=" * 80)
print("🎯 PREVIEW (Products with their sentiments):")
print("=" * 80)
df[['product_name', 'Rating', 'Customer_Review', 'Sentiment_Label', 'Sentiment_Polarity']].head(5)
```

```python
# 💡 BONUS: Quick insights from sentiment analysis
print("=" * 80)
print("💡 SENTIMENT ANALYSIS - KEY INSIGHTS")
print("=" * 80)

# 1. Category-wise sentiment
print("\n📊 SENTIMENT BY CATEGORY:")
print("-" * 50)
category_sentiment = df.groupby('Category')['Sentiment_Polarity'].mean().sort_values(ascending=False)
for cat, polarity in category_sentiment.head(10).items():
    if polarity > 0.3:
        emoji = "🟢"
    elif polarity > 0.1:
        emoji = "🟡"
    else:
        emoji = "🔴"
    print(f"{emoji} {cat:30} Avg Polarity: {polarity:.4f}")

# 2. Vendor-wise sentiment
print("\n🏪 SENTIMENT BY VENDOR:")
print("-" * 50)
vendor_sentiment = df.groupby('vendors')['Sentiment_Polarity'].mean().sort_values(ascending=False)
for vendor, polarity in vendor_sentiment.items():
    if polarity > 0.3:
        emoji = "🟢"
    elif polarity > 0.1:
        emoji = "🟡"
    else:
        emoji = "🔴"
    count = len(df[df['vendors'] == vendor])
    print(f"{emoji} {vendor:15} Polarity: {polarity:.4f} ({count:,} products)")

# 3. Sentiment vs Rating correlation
print("\n⭐ RATING vs SENTIMENT RELATIONSHIP:")
print("-" * 50)
rating_sentiment = df.groupby(pd.cut(df['Rating'], bins=[0, 2, 3, 4, 5]))['Sentiment_Polarity'].mean()
for rating_range, polarity in rating_sentiment.items():
    print(f"   Rating {rating_range}: Avg Sentiment: {polarity:.4f}")

# 4. Top positive and negative products
print("\n🏆 TOP 5 MOST POSITIVE PRODUCTS:")
print("-" * 50)
top_positive = df.nlargest(5, 'Sentiment_Polarity')[['product_name', 'Category', 'Rating', 'Sentiment_Polarity']]
print(top_positive.to_string(index=False))

print("\n😞 TOP 5 MOST NEGATIVE PRODUCTS:")
print("-" * 50)
top_negative = df.nsmallest(5, 'Sentiment_Polarity')[['product_name', 'Category', 'Rating', 'Sentiment_Polarity']]
print(top_negative.to_string(index=False))

# 5. Summary
print("\n" + "=" * 80)
print("🎯 QUICK SUMMARY:")
print("=" * 80)
print(f"✅ Total Reviews Analyzed: {len(df):,}")
print(f"🟢 Positive Reviews: {len(df[df['Sentiment_Label']=='POSITIVE']):,} ({len(df[df['Sentiment_Label']=='POSITIVE'])/len(df)*100:.2f}%)")
print(f"🔴 Negative Reviews: {len(df[df['Sentiment_Label']=='NEGATIVE']):,} ({len(df[df['Sentiment_Label']=='NEGATIVE'])/len(df)*100:.2f}%)")
print(f"📊 Average Sentiment: {df['Sentiment_Polarity'].mean():.4f}")
print(f"🏆 Best Vendor: {vendor_sentiment.index[0]} (Polarity: {vendor_sentiment.values[0]:.4f})")
print(f"📁 Best Category: {category_sentiment.index[0]} (Polarity: {category_sentiment.values[0]:.4f})")

print("\n🎉 SENTIMENT ANALYSIS COMPLETE!")
print("🚀 Ready for ML Model Training!")
```
